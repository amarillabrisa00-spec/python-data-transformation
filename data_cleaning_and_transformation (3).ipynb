## 2. Exploración inicial

## 3. Detección de valores faltantes

Se identifican datos nulos o incompletos presentes en el conjunto de datos.

```python
df.count() #revisa cuantos valores nulos hay en cada columna.
```

## 4. Tratamiento de valores nulos

Los valores faltantes son reemplazados para garantizar consistencia en el análisis.

```python
df = df.fillna(0) #reemplaza todos los valores vacios (nulos o NAN) en la tabla df por ceros

```

```python
df.count() #valores missing ya completados 
```

```python
print(df.columns) #muestra todas las columnas
```

## 5. Eliminación de columnas irrelevantes

Se eliminan variables no necesarias para optimizar el procesamiento del dataset.

```python
columns_to_drop = ['review_language','user_age','user_country']# borra tres columnas de la tabla df
df.drop(columns_to_drop, axis="columns", inplace=True) #inplace=True: hace el cambio directamente en df, sin crear una copia
```

```python
print(df.isna().sum()) #te muestra cuantos valores faltan (vacios o NAN) hay en cada columna de la tabla df.
```

## 6. Conclusión

El dataset fue limpiado y transformado correctamente, quedando preparado para análisis posteriores.
